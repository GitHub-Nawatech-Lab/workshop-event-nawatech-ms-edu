```python
# ATTENTION: AI-generated code can include errors or operations you didn't intend. Review the code in this cell carefully before running it.

df = spark.read.table('kemendikbud_lakehouse.dbo_data_sekolah').toPandas()
# df.head()
```

```python
df.drop("No", axis=1, inplace=True)
# df.head()
```

```python
def rename_columns(df):
    result = {col: f"{col.split('_')[0]}_{'Jumlah' if 'Jml' in col else 'Negeri' if 'N' in col else 'Swasta'}" for col in df.columns if '_' in col}
    return df.rename(columns=result)

def preprocess_column(column):
    if column.dtype == 'float64':
        column = column.astype('int64')
    return column

df = rename_columns(df)
# for column in df.columns:
#     df[column] = preprocess_column(df[column])
```
